## Question 1 - Extracting Tesla Stock Data Using yfinance - 2 Points

```python
import yfinance as yf
import pandas as pd

tesla_stock = yf.Ticker("TSLA")
tesla_stock_data = tesla_stock.history(period="max")
tesla_stock_data.head()
```

## Question 2 - Extracting Tesla Revenue Data Using Webscraping - 1 Points

```python
import requests
from bs4 import BeautifulSoup

tesla_url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'
tesla_html = requests.get(tesla_url).text
tesla_soup = BeautifulSoup(tesla_html, 'html.parser')
tesla_revenue_table = tesla_soup.find_all('table')[1]
tesla_revenue_data = pd.read_html(str(tesla_revenue_table))[0]

tesla_revenue_data.head()
```

## Question 4 - Extracting GameStop Revenue Data Using Webscraping - 1 Points

```python
gamestop_url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'
gamestop_html = requests.get(gamestop_url).text
gamestop_soup = BeautifulSoup(gamestop_html, 'html.parser')
gamestop_revenue_table = gamestop_soup.find_all('table')[1]
gamestop_revenue_data = pd.read_html(str(gamestop_revenue_table))[0]

gamestop_revenue_data.head()
```

```python
tesla_dashboard = pd.concat([tesla_stock_data['Close'], tesla_revenue_data['Tesla Quarterly Revenue(Millions of US $).1']], axis=1)
tesla_dashboard.columns = ['Stock Price', 'Revenue']
tesla_dashboard.plot(title='Tesla Stock Price vs Revenue')
```
